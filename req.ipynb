```python
import json
from pprint import pprint
import http.client
import urllib.parse

```

```python
patterns = {
    'VERB': 'pattern {V [upos=VERB]}',
    'VERB-o': 'pattern {V [upos=VERB];'
              'V -[obj]-> O}',
    'VERB-oi': 'pattern {V [upos=VERB];'
               'V -[iobj]-> I}',
}

corpora = [
    'UD_French-FQB@2.12',
    'UD_French-GSD@2.12',
    'UD_French-PUD@2.12',
    'UD_French-ParTUT@2.12',
    'UD_French-Sequoia@2.12',
    'UD_French-ParisStories@2.12',
    'UD_French-Rhapsodie@2.12',
    'UD_Old_French-SRCMF@2.12'
]
```

```python
conn = http.client.HTTPSConnection("gmb.marceau-h.fr")

payload = """{"pattern":"pattern { N [upos=NUM] }","corpus":"UD_French-GSD@2.12","lemma":true,"upos":true,"xpos":false,"features":true,"tf_wf":false,"order":"init","context":false,"clust1":"no","clust2":"no"}"""

payload = f"param={urllib.parse.quote(payload)}"

print(payload)

headers = {
    'content-type': "application/x-www-form-urlencoded",
}

conn.request("POST", "/search", payload, headers)

res = conn.getresponse()
data = res.read()

uuid = json.loads(data.decode("utf-8"))["data"]["uuid"]
uuid
```
